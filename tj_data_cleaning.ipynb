```python
import pandas as pd
import plotly
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import chart_studio.plotly as py
import plotly.graph_objs as go
import numpy as np
import re
```

```python
df_stores = pd.read_csv("scraped_store_data.csv")
df_fips = pd.read_csv("all-geocodes-v2018.csv")
df_fips = df_fips.drop(columns=['Summary Level','County Subdivision Code (FIPS)','Place Code (FIPS)','Consolidtated City Code (FIPS)'])
df_fips['city_name'] = df_fips['Area Name (including legal/statistical area description)']
df_fips = df_fips.drop(columns='Area Name (including legal/statistical area description)')
```

```python
# keeping only CA counties
df_fips_counties = df_fips[df_fips['State Code (FIPS)']==6]
df_fips_counties = df_fips_counties[df_fips_counties.iloc[:,2].str.contains("County$")]
df_fips_counties['city_name'] = df_fips_counties.iloc[:,2]
df_fips_counties.drop(columns='Area Name (including legal/statistical area description)')
```

```python
df_stores['city_name'] = df_stores['city_name'].str.lower()
df_fips['city_name'] = df_fips['city_name'].str.lower()
df_fips['city_name'] = df_fips['city_name'].apply(lambda x: re.sub("(\scity)$", "", x))
```

```python
df = pd.merge(df_stores, df_fips, how='left', on='city_name')
df = df[df["State Code (FIPS)"]== 6.0]
df_fips_counties.columns
```
